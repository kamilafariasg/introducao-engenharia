# **Derivada**


O conceito de derivada está intimamente relacionado à **taxa de variação instantânea** de uma função, o qual está presente no cotidiano das pessoas, através, por exemplo, da determinação da **taxa de crescimento de uma certa população**, **da taxa de crescimento econômico do país**, da **taxa de redução da mortalidade infantil**, da **taxa de variação de temperaturas, da velocidade de corpos ou objetos em movimento**, enfim, inúmeros exemplos. Para entendermos como isso se dá, inicialmente vejamos a definição matemática da derivada de uma função em um ponto

**Definição**: Se uma função $f$ é definida em um intervalo aberto contendo $x0$, então a derivada de $f$ em $x0$, denotada por  $f ’(x0)$, é dada por: 

$f'(x0) = \lim_{\Delta x \rightarrow 0}=  \frac{f(x0 + \Delta x)-f(x0)}{\Delta x} $

se este limite existir. $\Delta x $ representa uma pequena variação em $x$, próximo de $x0$, ou seja, tomando $\Delta x= x - x0$, a derivada de $f$ em  $x0$ pode também se expressa por:

$ f'(x0) = \lim_{\Delta x \rightarrow 0} \frac{f(x) - f(x0)}{x - x0}$

# **Resolução de Exercícios**


01. Calcule a derivada da função:
$f(x) = x^{10}$

```python
import sympy as sy
from sympy.interactive import init_printing
init_printing(pretty_print=True)

# Declara a variável simbólica
x = sy.Symbol('x')
# Explicita a função
funcao = x**10

# Resolve a função
resultado = sy.diff(funcao)
resultado
```

2. Calcule a derivada da função: $f(x)= -x^5+3x$

```python
import sympy as sy
from sympy.interactive import init_printing
init_printing(pretty_print=True)

# Declara a variável simbólica
x = sy.Symbol('x')
# Explicita a função
funcao = -x**5 + 3*x

# Resolve a função
resultado = sy.diff(funcao)
resultado
```

2. Nesse segundo exemplo será calculada a derivada parcial da equação abaixo em relação a x:

$ f = x^{10}y^{3}z^{8}$

```python
import sympy as sy
from sympy.interactive import init_printing
init_printing(pretty_print=True)

# Declara as variaveis simbolicas
x, y, z = sy.symbols('x y z')

# Funcao a ser resolvida
funcao = x**10 * y**3 * z**8

# Calcula o resultado da derivada parcial em funcao de X
resultadoRelativoX = sy.diff(funcao, z)
resultadoRelativoX
```

# **Problemas de Otimização**

Otimizar é encontrar a melhor solução para seu problema, ou seja, a que mais lhe favore. Dessa forma, a derivada pode ser usada nesse processo através dos máximos e mínimos. 

Pode-se imaginar que um problema de otimização se trabalha muito em situações do nosso dia a dia, ele se encontra  em diversos feitos da humanidade ou da natureza, como por  exemplo:  Maximizar  áreas,  volumes  e  lucros  ou  minimizar  distância,  tempo  e custos.

## **Exemplo 1:** 

Vamos supor que as visitas que um parque recebe variam de acordo com a temperatura, em ºC, de acordo com a função:

$V(t) = - t^2+50t $

Dessa forma, para qual temperatura o parque terá maior quantidade de visitantes?

```python
import sympy as sy
from sympy.interactive import init_printing
init_printing(pretty_print=True)

# Declara a variável simbólica
t = sy.Symbol('t')
# Explicita a função
funcao = (-t**2)+50*t

# Resolve a função
resultado = sy.diff(funcao)
print('A função derivada para solucionar a questão é:')
resultado
```

```python
from sympy.solvers import solve
print('O valor x da fução derivada é:')
solve(resultado,t)
```

## **Exemplo 2: Lucro máximo a partir de receita e custo** 

A Cia X Ltda produz determinado produto e vende-o a um preço unitário de R\$ 13. Estima-se que o custo total c para produzir e vender q unidades é dado por

 $ c = q^3 - 3q^2 + 4q + 2 $
 
Supondo que toda produção seja absorvida pelo mercado consumidor, qual é a quantidade que deverá ser produzida para se ter lucro máximo?

A solução da questão é dada pela equação:  $ c = 13q - (q^3 - 3q^2 + 4q + 2) ou -q^3+3q^2+9q-2$

```python
import sympy as sy
from sympy.interactive import init_printing
init_printing(pretty_print=True)

# Declara a variável simbólica
x = sy.Symbol('x')
# Explicita a função
funcao = -(x**3) + (3*(x**2)) + (9*x) - 2

# Resolve a função
resultado2 = sy.diff(funcao)
print('A função derivada para solucionar a questão é:')
resultado2
```

```python
print('O valor x da fução derivada é:')
solve(resultado2,x)
```

## **Exemplo 3: Maximizar lucro na produção**

Uma forma líquida de penicilina fabricada por uma firma farmacêutica é vendida a granel a um preço de R\$ 200 por unidade. Se o custo total de produção (em dólares) para x unidades for 

$ C(x) = 500.000 + 80x + 0,003x^2$

e se a capacidade de produção da firma for de, no máximo, 30.000 unidades em um tempo especificado, quantas unidades de penicilina devem ser fabricadas e vendidas naquele tempo para maximizar o lucro?

Solução: Para resolver este problema, temos que:

$ L = R - C$, onde $R = 200x$.

Assim, $ L = 200x -(500.000 + 80x + 0,003x^2)$. 

Simplificando, $ L = -0,003x^2 + 120x - 500.000$

```python
import sympy as sy
from sympy.interactive import init_printing
init_printing(pretty_print=True)

# Declara a variável simbólica
x = sy.Symbol('x')
# Explicita a função
funcao = -(0.003*(x**2)) + (120*x) - 500000

# Resolve a função
resultado3 = sy.diff(funcao)
print('A função derivada para solucionar a questão é:')
resultado3
```

```python
from sympy import *
print('O valor x da fução derivada é:')
solve(resultado3,x)
```

## **Exemplo 4: Dimensões**

Um certo cartaz deverá ter $600 cm^2$ para a mensagem impressa; deve ter 7,5 cm de margem no topo e na base e uma margem de 5 cm em cada lado. Determine as dimensões totais do cartaz para que a quantidade de papel usada seja mínima.

Solução: $A = (x + 10) (y+15)$, sabendo que $y = 600 / x$, temos :

$A = (x +10) (600/x + 15)$ . Simplificando,

$ A = 750 + 15x + 6000/x$

Derivando: 

```python
import sympy as sy
from sympy.interactive import init_printing
init_printing(pretty_print=True)

# Declara a variável simbólica
x = sy.Symbol('x')
# Explicita a função
funcao = 750 + 15*x + (6000/x)

# Resolve a função
resultado4 = sy.diff(funcao)
print('A função derivada para solucionar a questão é:')
resultado4
```

```python
from sympy import *
print('O valor x da fução derivada é:')
solve(resultado4,x)
```
